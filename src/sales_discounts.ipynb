#  Import Pandas Library

```python
import pandas as pd
```

# Load Sales DataFrame

```python
sales =  pd.read_csv('../data_input/sales.csv',header='infer')
```

# Show Sales data

```python
sales
```

# Load Discounts DataFrame

```python
discounts =  pd.read_csv('../data_input/discounts.csv',header='infer')
```

# Shows Discounts Data

```python
discounts
```

# Merge Sales with Discounts DataFrame

```python
sales_discounts  = sales.merge(discounts, left_on=['sales_order_id','customer_id'], right_on = ['sales_order_id','customer_id'])
```

# Show Result with discount_value's column 

```python
sales_discounts
```

```python
sales_discounts.describe()
```

 # Save result as a csv

```python
sales_discounts.to_csv('../data_output/sales_discounts.csv', index=False) 
```
